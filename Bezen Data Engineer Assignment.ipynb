```python
data.columns
```

```python
#Removing $ symbol from all rows of column price_string
data['price_string']=data['price_string'].str.replace('[$]','')
```

```python
data
```

```python
#Now adding dollar to price string column in the beginning 
data['price_string'] = "$" + data['price_string'].dropna().astype(str) 
```

```python
data
```

```python
#Repeating same thing with price_string_unf column
data['price_string_unf']=data['price_string_unf'].str.replace('[$]','')
data['price_string_unf'] = "$" + data['price_string_unf'].dropna().astype(str) 
data
```

```python
#Replacing dollar with $_ to split the data in two columns
data['price_string']=data['price_string'].str.replace('[$_A_]','')
data['price_string'] = "$_" + data['price_string'].dropna().astype(str) 
data
```

```python
#Splitting into two columns-value,currency
data[['Currency','Value']] = data['price_string'].str.split("_",expand=True)
```

```python
data
```

```python
#Adding dollar back to price_string column
data['price_string']=data['price_string'].str.replace('[$_]','')
data['price_string'] = "$" + data['price_string'].dropna().astype(str) 
data
```

```python
# Exporting this dataframe to CSV
data.to_csv("Q3.csv")
```
